```python
from langchain_community.utilities import SQLDatabase
db_uri = "mysql+mysqlconnector://root:SQLpassword5@localhost:3306/chinook"
db= SQLDatabase.from_uri(db_uri)
```

CRIAR função que retoma o schema da database

```python
def get_schema(_):
    return db.get_table_info()
```

```python
from langchain_core.output_parsers import StrOutputParser
from langchain_core.runnables import RunnablePassthrough
from langchain_openai import ChatOpenAI

llm=ChatOpenAI()

#o runnablepassthrough requer pelo menos um parametro e é por isso que 
#tive de definir def get_schema(_): com um underscore para garantir pelo menos um paraemtro
sql_chain = (
    RunnablePassthrough.assign(schema=get_schema)
    | prompt
    | llm.bind(stop="\nSQL Result:") #adicionar stop parameter(boa pratica) para que a llm pare de gerar depois de ter atingido o resultdo desejado
    | StrOutputParser() #para que o resultado desta SQL chain seja apenas a SQL query
)
```

```python
sql_chain.invoke({"question":"how many artists are there?"})
```

Criar a outer CHAIN

```python
template = """Based on the table schema below, question, sql query, and sql response, write a natural language response:
{schema}

Question: {question}
SQL Query: {query} #gerada pela sql chain
SQL Response: {response}"""
prompt_response = ChatPromptTemplate.from_template(template)
```

```python
def run_query(query):
    return db.run(query)
```

```python
run_query("SELECT COUNT(DISTINCT ArtistId) AS TotalArtists\nFROM album;")
```

```python
full_chain = (
    RunnablePassthrough.assign(query=sql_chain).assign(
        schema=get_schema,
        response=lambda vars: run_query(vars["query"]),
    )
    | prompt_response
    | llm
)
```

```python
full_chain.invoke({"question": "how many artists are there?"})
```
